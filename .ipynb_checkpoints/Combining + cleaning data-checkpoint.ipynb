```python
reviews_df = pd.read_csv('/Users/mahmoud/Car_Price_Prediction/car_reviews.csv')
```

```python
reviews_df = reviews_df.drop(columns=['Unnamed: 0'])
```

```python
car_data_df = pd.read_csv('/Users/mahmoud/Car_Price_Prediction/car_data_cleaned.csv')
```

```python
car_data_df = car_data_df.drop(columns=['Unnamed: 0'])
```

```python
reviews_df
```

```python
car_data_df
```

```python
# Having the mileage column in float rather than string
car_data_df['Mileage'] = car_data_df['Mileage'].str.replace(' mi.', '').str.replace(',', '')
car_data_df['Mileage'] = pd.to_numeric(car_data_df['Mileage'], errors='coerce')
car_data_df['Mileage']
```

```python
# Going from all the info of the car in 'Name' seperate them and create new columns
reviews_df[['Year', 'Make', 'Model', 'Type']] = reviews_df['Car Name'].str.extract(r'(\d{4})\s+([^ ]+)\s+([^ ]+)\s+(.*)')
reviews_df
```

```python
reviews_df.isna().sum()
```

```python
# Step 1: Function to extract 'Year', 'Make', and 'Model' from 'Car Name' column in 'reviews_df'
def extract_info_from_car_name(car_name):
    year, make_model = car_name.split(' ', 1)
    make, model = make_model.split(' ', 1)
    return year, make, model

# Step 2: Apply the function to extract info from 'Car Name' column and create new columns in 'reviews_df'
reviews_df['Year'], reviews_df['Make'], reviews_df['Model'] = zip(*reviews_df['Car Name'].map(extract_info_from_car_name))

# Step 3: Convert 'Year' column in 'reviews_df' to integer type
reviews_df['Year'] = reviews_df['Year'].astype(int)

# Step 4: Drop the 'Car Name' column from 'reviews_df' as it's no longer needed
reviews_df.drop(columns=['Car Name'], inplace=True)

# Step 5: Merge 'car_data_df' into 'reviews_df' based on 'Year', 'Make', and 'Model' to create 'merged_df'
merged_df = pd.merge(reviews_df, car_data_df[['Year', 'Make', 'Model']], on=['Year', 'Make', 'Model'], how='left')

# Step 6: Print the final merged dataframe with 'Mileage' added
merged_df
```
